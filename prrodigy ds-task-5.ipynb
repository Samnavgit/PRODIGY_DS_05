```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\lenovo\Desktop\Road.csv")
df
```

```python
print("Missing values in the dataset:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)  
print(df.isnull().sum())
     
```

```python
print("Data types and basic statistics:")
print(df.info())
```

```python
print(df.describe())
     
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_week', data=df)
plt.title('Accidents by Day of the Week')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex_of_driver', data=df)
plt.title('Accidents by Gender of the Driver')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Driving_experience', data=df)
plt.title('Accidents by Driving Experience of the Driver')
plt.show()
```

```python
df['Time'] = pd.to_datetime(df['Time']).dt.time
```

```python
plt.figure(figsize=(20, 8))
ax = sns.countplot(x='Time', data=df, order=sorted(df['Time'].unique()))
ax.set_xticks(ax.get_xticks()[::10])  # Show every 10th tick
plt.title('Accidents by Time of Day')
plt.xticks(rotation=90)
plt.show()
     
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Select only numeric columns
numeric_columns = df.select_dtypes(include=[np.number])

# Correlation Matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Calculate the average severity by weather condition
average_severity_by_weather = df.groupby('Weather_conditions')['Accident_severity'].value_counts(normalize=True).unstack().mean(axis=1)

# Display the result
print("Average Severity by Weather Condition:")
print(average_severity_by_weather)
     
```
